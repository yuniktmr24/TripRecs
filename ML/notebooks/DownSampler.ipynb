### All images in a directory resize

```python
from PIL import Image
import os

def resize_image_by_percentage(input_path, output_path, percentage):
    """Resize image by a percentage and save to output path."""
    with Image.open(input_path) as img:
        width, height = img.size
        new_dimensions = (int(width * percentage / 100), int(height * percentage / 100))
        resized_img = img.resize(new_dimensions, Image.Resampling.LANCZOS)
        resized_img.save(output_path)

def resize_images_in_directory(source_dir, target_dir, percentage):
    """Resize all images in the source directory and save them to the target directory."""
    # Create the target directory if it does not exist
    if not os.path.exists(target_dir):
        os.makedirs(target_dir)
    
    for filename in os.listdir(source_dir):
        # Construct the full file path
        try:
            input_path = os.path.join(source_dir, filename)
            output_path = os.path.join(target_dir, filename)
            
            # Check if the file is an image (you might want to add other formats)
            if input_path.lower().endswith(('.png', '.jpg', '.jpeg', '.tiff', '.bmp', '.gif')):
                resize_image_by_percentage(input_path, output_path, percentage)
                print(f"Resized and saved: {output_path}")
            else:
                print(f"Skipped non-image file: {filename}")
        except:
            print(f"Error processing:  {filename}")

# Example usage:
folder_name_list = ['53.Toronto_Harbour', '54.Pyrohiv', '55.Harvard']

for folder_name in folder_name_list:
    #for train directory
    source_directory = '../landmark_images/train/' + folder_name
    target_directory = '../resize/train/' + folder_name
    
    percentage = 35  # Resize to 50% of the original size
    
    resize_images_in_directory(source_directory, target_directory, percentage)
    
    #for test directory
    source_directory = '../landmark_images/test/' + folder_name
    target_directory = '../resize/test/' + folder_name
    
    resize_images_in_directory(source_directory, target_directory, percentage)
```
