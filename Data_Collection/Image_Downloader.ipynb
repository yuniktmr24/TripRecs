### Download TIFs as well with on the fly JPG conversion

```python
import os
import requests
import hashlib
from PIL import Image
import io

def download_image(image_url, folder_path):
    """Download an image and convert .tif to .jpg if necessary before saving."""
    headers = {'User-Agent': 'your-email@example.com'}
    response = requests.get(image_url, headers=headers, stream=True)

    if response.status_code == 200:
        # Create a unique filename using the URL's hash
        url_hash = hashlib.md5(image_url.encode('utf-8')).hexdigest()
        # Determine the file extension (and subsequently, the format)
        if image_url.lower().endswith('.tif') or image_url.lower().endswith('.tiff'):
            format = 'JPEG'  # We'll convert TIFF images to JPEG
            extension = 'jpg'
        else:
            format = 'JPEG'  # Default to JPEG for other formats; adjust if necessary
            extension = 'jpg'
        
        filename = f"{url_hash}.{extension}"
        file_path = os.path.join(folder_path, filename)

        if format == 'JPEG' and not image_url.lower().endswith('.jpg'):
            # If the image needs to be converted to JPEG
            try:
                image = Image.open(io.BytesIO(response.content))
                image = image.convert('RGB')  # Convert to RGB in case it's a different mode
                image.save(file_path, format=format)
                print(f"Image converted to JPEG and saved to {file_path}")
            except Exception as e:
                print(f"Failed to convert or save image: {e}")
        else:
            # If the image is already a JPEG, save directly
            with open(file_path, 'wb') as f:
                for chunk in response.iter_content(chunk_size=8192):
                    f.write(chunk)
            print(f"Image saved to {file_path}")
    else:
        print(f"Failed to download image: {image_url}")

def ensure_directory_exists(path):
    """Ensure that a directory exists; if not, create it."""
    if not os.path.exists(path):
        os.makedirs(path)
    return path
```

```python
from PIL import Image
import os

def convert_to_grayscale(source_directory, target_directory):
    # Ensure the target directory exists, create it if it doesn't
    os.makedirs(target_directory, exist_ok=True)
    
    # Loop through each file in the source directory
    for filename in os.listdir(source_directory):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):  # Check for image files
            file_path = os.path.join(source_directory, filename)
            # Open the image file
            with Image.open(file_path) as img:
                # Convert the image to grayscale
                grayscale_img = img.convert('L')
                # Save the grayscale image to the target directory
                grayscale_img.save(os.path.join(target_directory, f'grayscale_{filename}'))

# Replace 'source_folder_path' with the path to your folder containing images
# Replace 'target_folder_path' with the path where you want to save grayscale images
convert_to_grayscale('../ML/sample_inferences/regular', '../ML/sample_inferences/grayscale')
```

```python
from PIL import Image
import numpy as np
import os

def add_noise(image):
    # Convert the image to an array for noise addition
    image_array = np.array(image)
    # Generate noise and add it to the image array
    noise = np.random.randint(-25, 25, image_array.shape, dtype='int16')
    noisy_image = image_array + noise
    # Ensure values are within the valid range [0, 255]
    noisy_image = np.clip(noisy_image, 0, 255)
    # Convert array back to an image
    noisy_image = Image.fromarray(noisy_image.astype('uint8'))
    return noisy_image

def add_noise_to_images(source_directory, target_directory):
    # Ensure the target directory exists, create it if it doesn't
    os.makedirs(target_directory, exist_ok=True)
    
    # Loop through each file in the source directory
    for filename in os.listdir(source_directory):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):  # Check for image files
            file_path = os.path.join(source_directory, filename)
            # Open the image file
            with Image.open(file_path) as img:
                # Add noise to the image
                noisy_image = add_noise(img)
                # Save the noisy image to the target directory
                noisy_image.save(os.path.join(target_directory, f'noisy_{filename}'))

# Replace 'source_folder_path' with the path to your folder containing images
# Replace 'target_folder_path' with the path where you want to save noisy images
add_noise_to_images('../ML/sample_inferences/regular', '../ML/sample_inferences/noisy')
```
